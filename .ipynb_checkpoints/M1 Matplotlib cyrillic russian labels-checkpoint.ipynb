## Matplotlib cyrillic labels. Matplotlib подписи по-русски.

При публикации результатов научной работы в отечественных журналах, при выступлениях на докладах, конференциях и презентациях перед родной аудиторией хорошим тоном считается подписывать рисунки на языке родной страны, то есть в моём случае на русском языке. В этом notebook разберёмся, как легко делать подписи рисунков в matplotlib на русском. Стоит отметить, что дальнейшая инструкция касается pyhton2 версии 2.7.1. и выше как для Windows (дистрибутив Python(x,y), так и для Linux (Ubuntu 12.04 LTS). В более ранних версиях python, возможно, могут быть отличия от представленных ниже схем.

#### Настройка 1. Шрифты

Основная проблема matplotlib для отображения кириллических символов состоит в том, что по умолчанию используются шрифты, которые не поддерживают кириллицу. Соответственно, необходимо сменить шрифт на тот, который поддерживает кириллицу. В ОС Windows это может быть, например, Arial, Times New Roman или Tahoma. В ОС Linux Ubuntu 12.04 LTS - Ubuntu. Сделать это можно как в теле самой программы-скрипта, так и в профильных настройках.


C помощью matplotlib.rcParams можно, во-первых, установить все параметры рисования (шрифты, толшина/тип линий, оформление рамок) на необходимые значения, а во-вторых, увидеть, собственно, список настроек, то есть список того, что можно изменить.

```python
import matplotlib as mpl
mpl.rcParams # показывает текущие настройки
mpl.rcParamsDefault # показывает значения параметров, которые хранятся в файле matplotlibrc
mpl.rcdefaults() # сбрасывает значения на те, которые хранятся в файле matplotlibrc
```

Сами настройки лежат в файле с названием matplotlibrc. Чтобы узнать какой именно файл используется (то есть где он будет искаться системой) есть специальная функция:

```python
import matplotlib as mpl
mpl.matplotlib_fname() # показывает откуда был загружен текуший файл matplotlibrc
```

Дополнительную информацию по использованию этого файла есть здесь:

http://matplotlib.org/users/customizing.html

Там же есть и  типичный файл matplotlibrc, если он вдруг испортился.

```python
import matplotlib as mpl
mpl.rcParams # показывает текущие настройки
```

Из настроек matplotlib нас интересуют fonts, то есть шрифты. Есть 5 стилей шрифтов: cursive, fantasy, monospace, sans-serif и serif. Ещё есть family шрифт, который определяет, какой из пяти стилей будет использоваться по умолчанию. В приведенном выше примере будет использоваться sans-serif.
У каждого стиля шрифта определены списки из реальных шрифтов, которыми пользуются в текстовых процессорах и редакторах. Чтобы поставить шрифт, который будет корректно отображать кириллицу, можно воспользоваться таким примером:

```python
import matplotlib as mpl
mpl.rcParams['font.family'] = 'fantasy'
mpl.rcParams['font.fantasy'] = 'Arial' # Для Windows
mpl.rcParams['font.fantasy'] = 'Ubuntu' # Для Ubuntu
mpl.rcParams['font.fantasy'] = 'Arial, Ubuntu' # работает как под Windows, так и под Ubuntu
```

Таким образом, стиль fantasy будет основным, и из него будут выбираться шрифты Arial или Ubuntu, есл первый не будет найден. Оба эти шрифта корректно отображают кирилличекие символы.

Можно использовать и такой синтаксис:

```python
from matplotlib import rc
mpl.rcdefaults() # сброс настроек  

font = {'family': 'Courier New',
        'weight': 'normal'}
rc('font', **font)

plt.title(u'Изменился шрифт всех подписей!')
plt.show()
```

Или такой:

```python
mpl.rcdefaults() # сброс настроек 
plt.xlabel(u'Ось абсцисс',{'fontname':'Arial'})
plt.ylabel(u'Ось ординат',{'fontname':'Times New Roman'})
plt.title(u'Заголовок и буква Ё! А также разные шрифты элементов рисунка',{'fontname':'Tahoma'})
plt.show()
```

Список доступных в ОС шрифтов можно узнать, например, вот так:

```python
import matplotlib.font_manager as fm
for i in (sorted(fm.findSystemFonts())):
    print i
```

Для поиска по названию известного вам шрифта в ОС можно воспользоваться функцией findfont:

```python
import matplotlib.font_manager as fm
print 'Courier New - %s' % fm.findfont('Courier New') # Ищем Courier New
print 'Times New Roman - %s' % fm.findfont('times new roman') # Ищем times new roman строчными буквами
print 'Comic Sans MS - %s' % fm.findfont('COMIC SANS MS') # Ищем COMIC SANS MS прописными буквами
```

N.B. Для разных Linux дистрибутивов, где может не быть известных вам шрифтов, корректно отображающих кириллицу, можно подобрать один из шрифтов из получившегося списка. Форумы и поисковики могут сильно помочь в этом. Для Ubuntu 12.04 LTS, например, есть шрифт Ubuntu. Для Ububntu 12.04 LTS список шрифтов будет выглядеть примерно так:

```python
import matplotlib.font_manager as fm
for i in (sorted(fm.findSystemFonts())):
    print i
```

```python
import matplotlib.font_manager as fm
print 'Ubuntu - %s' % fm.findfont('Ubuntu') # Ищем Ubuntu
print 'Times New Roman - %s' % fm.findfont('times new roman') # Ищем times new roman строчными буквами
print 'Comic Sans MS - %s' % fm.findfont('COMIC SANS MS') # Ищем COMIC SANS MS прописными буквами
```

#### Настройка 2. Юникод

Вторым необходимым условием для отображения на рисунках matplotlib русских букв является использование unicode-строк. В python3 все строки являются юникодовыми, там дополнительных примочек не нужно. В python2 необходимо преобразовывать обычную строку в юникод-строку: 

```python
str1 = 'строка 1' # обычная строка
print str1,type(str1)
str2 = u'строка 1' # unicode-строка
print str2,type(str2)
```

Также ставим в начале скрипта: # -*- coding: utf-8 -*- . В GUI Spyder такая запись автоматически вставляется в новый созданный скрипт. Эта GUI входит в состав дистрибутива winpython и python(x,y) для Windows.  

Суммируем всё выше изложенное и приведём пример.

#### Пример

```python
# -*- coding: utf-8 -*-

from math import pi
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcdefaults() # cбрасываем настройки на "по умолчанию"
mpl.rcParams['font.family'] = 'fantasy'
mpl.rcParams['font.fantasy'] = 'Times New Roman', 'Ubuntu','Arial','Tahoma','Calibri'

x = np.arange(-pi,pi+0.1,0.1)

fig = plt.figure()
ax1 = fig.add_subplot(111)
y = np.tanh(x)
y2 = np.arctan(x)
ax1.plot(x,y,'k--',label=u'Гиперболический тангенс')
ax1.plot(x,y2,'b',label=u'Арктангенс')

st = u'Заголовок по-русски'
sy = u'Ось ординат OY по-русски'
sx = u'Ось абсцисс OX по-русски'

ax1.set_title(st,size=20,color='green')
ax1.set_xlabel(sx,size=14,color='orange')
ax1.set_ylabel(sy,size=12, color='black')
ax1.grid(True)
ax1.legend(loc='best',frameon=False)

plt.tight_layout() # автоматическое выравнивание элементов на холсте plt
plt.show()
```

#### P.S. Python3 и русские подписи

Для python3 все строки являются юникодовыми по умолчанию. Значит, можно не указывать перед строками конвертер u. В остальном ситуация такая же - необходимо настроить шрифты, которые корректно отображают русские буквы/кириллицу.
